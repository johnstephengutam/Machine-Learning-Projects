```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Generate some synthetic training data
X_train = np.array([[1], [2], [3], [4], [5]])  # Feature vector (input)
y_train = np.array([0, 0, 1, 1, 1])           # Target variable (output)

# Create polynomial features
degree = 2
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X_train)

# Create and train a linear regression model
model = LinearRegression()
model.fit(X_poly, y_train)

# Generate test data
X_test = np.linspace(0, 6, 100).reshape(-1, 1)  # Generate 100 evenly spaced points between 0 and 6

# Transform test data using polynomial features
X_test_poly = poly_features.transform(X_test)

# Make predictions
y_pred = model.predict(X_test_poly)

# Plot the training data and the learned curve
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.plot(X_test, y_pred, color='red', label='Learned Curve')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Non-linear Regression using Polynomial Regression')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid activation function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Gradient of the cross-entropy loss function
def gradient(X, y, w):
    y_pred = sigmoid(np.dot(X, w))
    return np.dot(X.T, (y_pred - y)) / len(y)

# Binary cross-entropy loss function
def loss(X, y, w):
    y_pred = sigmoid(np.dot(X, w))
    return -np.mean(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))

# Gradient descent algorithm
def gradient_descent(X, y, learning_rate=0.01, num_iterations=1000, tolerance=1e-3):
    # Initialize weights
    w = np.zeros(X.shape[1])
    losses = []
    
    for i in range(num_iterations):
        # Compute gradient
        grad = gradient(X, y, w)
        # Update weights
        w -= learning_rate * grad
        # Compute loss
        curr_loss = loss(X, y, w)
        losses.append(curr_loss)
        
        # Check for convergence
        if np.linalg.norm(grad) < tolerance:
            break
    
    return w, losses

# Load and preprocess data
# Assuming X_train, y_train, X_test, y_test are already loaded

# Add intercept term to X_train and X_test
X_train_bias = np.c_[np.ones((X_train.shape[0], 1)), X_train]
X_test_bias = np.c_[np.ones((X_test.shape[0], 1)), X_test]

# Train logistic regression model using gradient descent
weights, losses = gradient_descent(X_train_bias, y_train)

# Plot loss curve
plt.plot(losses)
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.show()

# Evaluate model on training set
train_pred = sigmoid(np.dot(X_train_bias, weights))
train_pred_class = np.where(train_pred >= 0.5, 1, 0)
train_accuracy = np.mean(train_pred_class == y_train)
print("Training Accuracy:", train_accuracy)

# Evaluate model on test set
test_pred = sigmoid(np.dot(X_test_bias, weights))
test_pred_class = np.where(test_pred >= 0.5, 1, 0)
test_accuracy = np.mean(test_pred_class == y_test)
print("Test Accuracy:", test_accuracy)
```
